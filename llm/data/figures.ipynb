```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Function to create a confusion matrix from TN, FP, FN, TP
def create_confusion_matrix(tn, fp, fn, tp):
    return np.array([[tn, fp], [fn, tp]])

# Define the TN, FP, FN, TP values for each confusion matrix
cm1 = create_confusion_matrix(tn=23, fp=2, fn=17, tp=8)
cm3 = create_confusion_matrix(tn=2, fp=23, fn=1, tp=24)
cm2 = create_confusion_matrix(tn=20, fp=5, fn=7, tp=18)

# Titles for the confusion matrices
titles = ['Zero-shot', 'Optimal (6 epochs)', 'Overfitting (10 epochs)']

# Plotting the confusion matrices side by side
fig, axes = plt.subplots(1, 3, figsize=(9, 3))

# Iterate over the confusion matrices and plot each one
for i, (cm, title) in enumerate(zip([cm1, cm2, cm3], titles)):
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, ax=axes[i], annot_kws={"size": 19})
    axes[i].set_title(title, fontsize=14)
    axes[i].set_xlabel('Predicted Label', fontsize=12)
    axes[i].set_ylabel('True Label', fontsize=12)
    axes[i].xaxis.set_ticklabels(['Natural', 'Artificial'])
    axes[i].yaxis.set_ticklabels(['Natural', 'Artificial'])

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
